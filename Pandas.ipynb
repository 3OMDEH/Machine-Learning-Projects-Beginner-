# **Introduction to Pandas**

image.png

**Pandas** is an open-source Python library designed for easy and efficient data manipulation and analysis.

Created by **Wes McKinney** in 2008, Pandas was developed to provide a high-performance tool for handling structured data in Python.

The name "Pandas" is derived from **"Panel Data"**, a term from econometrics, although the Panel data structure is no longer widely used in Pandas.

### Key Features of Pandas:
- **Data Loading**: Easily read data from different file formats such as CSV, Excel, JSON, and more into memory as a **DataFrame**.
- **Data Manipulation**: Perform operations such as inserting, updating, and deleting rows or columns within the DataFrame.
- **Missing Data Handling**: Efficiently manage missing or incomplete data with built-in tools.
- **Data Merging**: Combine multiple datasets with high performance.
- **Group By**: Aggregate and transform data by grouping based on specific criteria.
- **Data Exporting**: Export the processed data to various file formats like CSV, Excel, and others.

# **Pandas Data Strucure**

image.png

Pandas is built on top of **NumPy**, which gives it powerful performance and efficiency when handling large datasets. Pandas provides the following main data structures:

1. **Series**:
   - A one-dimensional (1D) array-like structure that holds homogeneous data, meaning all elements are of the same type.
   - The **size** of a Series is immutable (fixed), but the values within it can be changed (mutable).

2. **DataFrame**:
   - A two-dimensional (2D) tabular data structure, similar to a spreadsheet or SQL table, where each column can hold data of a different type (heterogeneous).
   - Both the **size** and the **data** in a DataFrame are mutable, allowing easy modification.

> **Note**: The **Panel** data structure, previously used for 3D data, has been deprecated in recent versions of Pandas. For working with higher-dimensional data (3D or more), its recommended to use **MultiIndex DataFrames** or external libraries like **xarray**.


# **Series**

image.png

A **Pandas Series** is a one-dimensional array-like object that can hold any data type, such as integers, strings, floats, or even Python objects. It is similar to a column in a table or a list in Python but comes with additional functionality.

### Key Features of a Pandas Series:
- **Homogeneous Data**: All elements in a Series must be of the same data type.
- **Labeled Index**: Each element in a Series is associated with an index label, allowing for easy access to individual elements.
- **Mutable Data**: While the values within a Series can be changed (mutable), the size (number of elements) is fixed once the Series is created.

# **Data-Frames**

image.png

A **Pandas DataFrame** is a two-dimensional, tabular data structure similar to an Excel spreadsheet or SQL table. It consists of rows and columns, where each column can hold different data types, such as integers, strings, floats, or objects.

### Key Features of a DataFrame:
- **Heterogeneous Data**: Different columns in a DataFrame can hold different data types.
- **Labeled Indexes**: Each row and column has an index label for easy data access.
- **Size Mutable**: Both the size and the data in a DataFrame can be modified (you can add/remove rows and columns).
- **Data Alignment**: DataFrame automatically aligns data by its labels when performing operations between DataFrames.

You can create a DataFrame using different inputs like dictionaries, lists, or other DataFrames.

```python
# Define a dictionary containing employee data
data = {'Name':['Jai', 'Anuj', 'Jai', 'Princi','Gaurav', 'Anuj', 'Princi', 'Abhi'],
		     'Age':[27, 24, 22, 32,33, 36, 27, 32],
		     'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
				 'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
		     'Qualification':['Msc', 'MA', 'MCA', 'Phd','B.Tech', 'B.com', 'Msc', 'MA'],
         'Salary' : [400, 600, 500, 600, 550, 370, 800, 1000]}


# Create Data-Frame from dictionary of data
df = pd.DataFrame(data)

print(df)
```

```python
# In order to split the data, we use groupby() function
# this function is used to split the data into groups based on some criteria
```

```python
df.groupby('Name').groups
```

```python
# group the data on Name value.
groupByName = df.groupby("Name")
```

```python
# first function return Data-Frame of first entries in all the groups formed
groupByName.first()
```

```python
# last function return Data-Frame of the last entries in all the groups formed
groupByName.last()
```

```python
# print each group name with data
for name, group in groupByName:
  print(name)
  print(group)
```

```python
groupByQualification = df.groupby("Qualification")
```

```python
for name, group in groupByQualification:
  print(name)
  print(group)
```

```python
#  get_group function retun Data-Frame based on group name
groupByQualification.get_group('MA')
```

Grouping data with multiple keys

```python
group = df.groupby(['Name', 'Qualification'])
```

```python
group.groups
```

```python
# return Data-Frame based on ('Princi', 'Msc') group name
group.get_group(('Princi', 'Phd'))
```

```python
df.groupby(['Qualification']).sum()
```

Applying function to a group

Aggregation is a process in which we compute a summary statistic about each group

```python
# 1- Aggregation
#  Aggregated function returns a single aggregated value for each group
```

```python
# performing aggregation using aggregate method
group_1 = df.groupby('Name')
```

```python
group_1.aggregate(np.sum)
```

```python
df.groupby(['Name', 'Qualification']).aggregate(np.sum)
```

```python
# Apply a multiple functions by passing a list of functions.
df.groupby(['Name']).aggregate([np.sum, np.mean, np.std])
```

```python
df.groupby(['Qualification']).aggregate([np.sum, np.mean, np.std])
```

```python
# Apply a different aggregation to the columns of a dataframe.
df.groupby(['Qualification']).aggregate({'Age' : np.sum, 'Salary' : [np.sum, np.mean]})
```

## **8- Function Application**

pipe : Table wise Function Application

apply : Row or Column Wise Function Application

applymap : Element wise Function Application

```python
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print(df)
```

```python
def adder(num1,num2):
   return num1+num2


def addFive(num):
  return num + 5
```
